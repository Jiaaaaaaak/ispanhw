```python
import requests

def search_costco(keyword, max_items=5):
    url="https://www.costco.com.tw/rest/v2/taiwan/products/search"


    params ={
        "fields":"FULL",
        "query":keyword,
        "pagesize":max_items,
        "lang":"zh_TW",
        "curr":"TWD"
    }

    headers = {"User-Agent":"Mozilla/5.0"}

    r = requests.get(url, params=params, headers=headers)
    data = r.json()

    products = data.get("products")or []

    if not products:
        return "查無此商品"
    

    results = []
    for p in products:
        results.append({
           "name": p.get("name"),
           "price": p.get("price", {}).get("value"),
           "url": "https://www.costco.com.tw" + p.get("url", ""),
           "image":"https:www.costco.com.tw" + p["image"][0]["url"]
           if p.get("images") else None
        })

    return results
```
